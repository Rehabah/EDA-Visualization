```python
df=sns.load_dataset('iris')
```

# sweetviz 

Provides a detailed report with visualization

```python
analysis_report=sv.analyze(df)
```

Displaying the report on webpage

```python
analysis_report.show_html('report.html', layout='vertical',scale=0.7)
```

Displaying the report on the same notebook

```python
analysis_report.show_notebook(w=1500, h=300,layout='widescreen')
```

# ProfileReport

Generating reports that includes (missing values, correlations, duplicated records, varibles distribution)

```python
data_profile=ProfileReport(df,title='Report')
```

Displaying the report as HTML page

```python
#data_profile
data_profile.to_file(output_file='Report1.html')
```

Displaying the report on the notebook

```python
data_profile.to_widgets()
```

```python
#or use
```

```python
data_profile.to_notebook_iframe()
```

Saving the report

```python
data_profile.to_file('report.json')
#data_profile.to_file('report.html')
```
